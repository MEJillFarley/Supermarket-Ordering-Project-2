```python
import pandas as pd
import datetime
import numpy as np
```

```python
#import csv files
invoice_df = pd.read_csv("Data_Folder/Invoices.csv")
order_leads_df = pd.read_csv("Data_Folder/OrderLeads.csv")
sales_df = pd.read_csv("Data_Folder/SalesTeam.csv")

```

```python
#display invoice and order_leads dataframes
invoice_df.head()
```

```python
order_leads_df.head()
```

```python
order_leads_df.isnull().sum()
```

```python
#split date into a separate year and month column 
order_leads_df["Year"] = order_leads_df["Date"].str.split('-', expand = True)[2]
order_leads_df["Month"] = order_leads_df["Date"].str.split('-', expand = True)[1]
order_leads_df.head()
```

```python
#convert months columns from month number to month name
month_dict = {"01" : 'January', "02" : 'February',
            "03" : 'March', "04": 'April', "05": 'May',
            "06": 'June', "07": 'July', "08": 'August',
            "09": 'September', "10": 'October', "11": 'November', "12": 'December'}
clean_df = order_leads_df.replace({"Month": month_dict})
clean_df.head()
```

```python
#groupby year, month, and percentage of converted orders
year_month_df = (clean_df.groupby(["Year", "Month"])["Converted"].sum())/(clean_df.groupby(["Year", "Month"])["Converted"].count())
new_df = pd.DataFrame([year_month_df])

#transpose dataframe 
year_month_clean = new_df.T

#rename converted column to conversion rate
year_month_clean1 = year_month_clean.rename(columns = {"Converted" : 'Conversion Rate'})
year_month_clean1.head()
```

```python
#group by year and company name, then calculate conversion percentage
company_name_groupby = clean_df.groupby(["Year", "Company Name"])["Converted"].sum()/clean_df.groupby(["Year", "Company Name"])["Converted"].count()
company_name_df = pd.DataFrame([company_name_groupby])

#transpose dataframe
company_name_df_clean = company_name_df.T

#rename converted column to conversion rate
company_name_df_clean1 = company_name_df_clean.rename(columns = {"Converted" : "Conversion Rate"})
company_name_df_clean1.head()
```
